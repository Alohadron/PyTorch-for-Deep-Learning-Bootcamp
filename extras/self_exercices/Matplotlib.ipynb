# Matplotlib

MAtplotlib is a popular Python library used for creatic static, animated, and interactive visualization. It provides an interface similar to MATLAB and works well with NumPy


### Common Plot Types

1. **Line Plot** (`plt.plot()`) - For continoues data.
2. **Scatter Plot** (`plt.scatter()`) - For discrete data points.
3. **Bar Chart** (`plt.bar()`) - For categorial data.
4. **Histogram** (`plt.hist()`) - For data distribution.
5. **Pie Chart** (`plt.pie()`) - For proportion-based data.

```python
# 1. Line Plot: Plot a simple line graph for y = 2x + 1 over the range x = [0, 10].
import matplotlib.pyplot as plt
import numpy as np

# Define x values
x = np.linspace(0, 10, 100) # 100 points between 0 and 10

# Compute y values
y = 2 * x + 1

# Create the plot
plt.plot(x, y, label='y = 2x + 1', color='blue', linestyle='-', linewidth=2)

# Add labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Line Plot: y = 2x + 1')

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

```python
# Scatter Plot: Create a scatter plot using 50 random x and y values (use numpy.random.rand).
import matplotlib.pyplot as plt
import numpy as np

# Generate random x and y values
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Create the scatter plot
plt.scatter(x, y, color='red', marker='o', edgecolors='black', alpha=0.7)

# Add labels and title
plt.xlabel("Random X values")
plt.ylabel("Random Y values")
plt.title("Scatter Plot of Random Points")

# Add grid
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
# Bar Chart: Plot a bar chart representing the population of five different cities.
import matplotlib.pyplot as plt

# Data for the bar chart
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
population = [8419600, 3980400, 2716000, 2328000, 1690000]

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(cities, population, color=['blue', 'green', 'red', 'purple', 'orange'])

# Add labels and title
plt.xlabel("Cities")
plt.ylabel("Population")
plt.title('Population of Five Different Cities')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
# Histogram: Generate a histogram for a dataset of 1000 normally distributed random numbers.
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset of 1000 normally distributed random numbers
data = np.random.randn(1000)

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Value')
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Random Numbers")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Pie Chart: Create a pie chart showing the percentage distribution of five different products.
import matplotlib.pyplot as plt

# Data
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
percentages = [25, 20, 15, 30, 10]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=products, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple', 'orange'], startangle=140)

plt.title('Percentage Distribution of Five Different Products')
plt.show()
```

```python
# Multiple Lines on the Same Plot: Plot y1 = sin(x), y2 = cos(x), and y3 = tan(x) on the same graph with different colors and line styles.
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x)', color='green', linestyle='--')
plt.plot(x, y3, label='tan(x)', color='red', linestyle=":")

# Set limits to avoid extreme values of tan(x)
plt.ylim(-5, 5)

# Add labels, title, and legend
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Plot of sin(x), cos(x), and tan(x)')
plt.legend()
plt.grid()

plt.show()
```

```python
# Subplots:
    # Create a figure with 2x2 subplots:
    # Top-left: Line plot
    # Top-right: Scatter plot
    # Bottom-left: Bar chart
    # Bottom-right: Histogram

# Customize a plot with a title, labels, grid, and a legend.

import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)
scatter_x = np.random.rand(50) * 10
scatter_y = np.random.rand(50)
cities = ['A', 'B', 'C', 'D', 'E']
population = [10, 20, 15, 25, 30]
data = np.random.randn(1000)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Top-left: Line plot
axs[0, 0].plot(x, y, color='blue', linestyle='-', label='sin(x)')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('sin(x)')
axs[0, 0].grid()
axs[0, 0].legend()

# Top-right: Scatter plot
axs[0, 1].scatter(scatter_x, scatter_y, color='red', label='Random Points')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('Random X')
axs[0, 1].set_ylabel('Random Y')
axs[0, 1].grid()
axs[0, 1].legend()

# Bottom-left: Bar chart
axs[1, 0].bar(cities, population, color=['blue', 'green', 'red', 'purple', 'orange'], label='Population')
axs[1, 0].set_title('Bar Chart')
axs[1, 0].set_xlabel('Cities')
axs[1, 0].set_ylabel('Population')
axs[1, 0].grid(axis='y', linestyle='--', alpha=0.7)
axs[1, 0].legend()

# Bottom-right: Histogram
axs[1, 1].hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7, label='Data Distribution')
axs[1, 1].set_title('Histogram')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].grid(axis='y', linestyle='--', alpha=0.7)
axs[1, 1].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Styled Plot: Use a specific Matplotlib style (e.g., ggplot, seaborn-darkgrid).
import matplotlib.pyplot as plt

# Use a valid Matplotlib style
plt.style.use('ggplot')

# Data
cities = ['A', 'B', 'C', 'D', 'E']
population = [10, 20, 15, 25, 30]

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(cities, population, color=['blue', 'green', 'red', 'purple', 'orange'], label='Population')

# Customize the plot
plt.title('City Population')
plt.xlabel("Cities")
plt.ylabel("Population")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Dual Y-Axis: Create a plot with two y-axes, one showing y1 = x^2 and the other showing y2 = sqrt(x).
import matplotlib.pyplot as plt
import numpy as np

# Matplotlib style
plt.style.use('Solarize_Light2')

# Generate data
x = np.linspace(1, 10, 100)
y1= x ** 2
y2 = np.sqrt(x)

# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot y1= x^2
ax1.plot(x, y1, color='blue', label='y = x^2')
ax1.set_xlabel('X values')
ax1.set_ylabel('y = x^2', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Create the second y-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, color='red', linestyle='--', label='y = sqrt(x)')
ax2.set_ylabel('y = sqrt(x)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add title and legend
plt.title('Dual Y-Axis Plot: y = x^2 and y = sqrt(x)')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.show()
```

```python
# 3D Plot: Create a 3D surface plot of z = sin(sqrt(x^2 + y^2)).
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z based on the given function
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
# Heatmap: Generate a heatmap using imshow() with a random 10x10 matrix.
import numpy as np
import matplotlib.pyplot as plt

# Generate a random 10x10 matrix
data = np.random.rand(10, 10)

# Create a heatmap using imshow
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.colorbar() # Show color scale
plt.title('Heatmap of Random 10x10 Matrix')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('classic')

# For Jupyter Notebooks, use this line to ensure the animation works
%matplotlib inline

# Create the figure and axis
fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 100)  # X-axis values

# Set the axis limits
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)

# Create a line object for the sine wave
line, = ax.plot([], [], lw=2)

# Initialize the line for the animation (empty at the start)
def init():
    line.set_data([], [])
    return line,

# Update function for the animation
def update(frame):
    line.set_data(x[:frame], np.sin(x[:frame]))  # Update the sine wave
    return line,

# Create the animation using FuncAnimation
ani = FuncAnimation(fig, update, frames=len(x), init_func=init, blit=True, interval=50)

# Display the animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)  # Create a mesh grid for the coordinates

# Define the function z = x^2 - y^2
Z = X**2 - Y**2

# Create a contour plot
plt.contour(X, Y, Z, 20, cmap='viridis')  # 20 contour levels, using 'viridis' colormap

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour plot of z = x^2 - y^2')

# Show the plot
plt.colorbar()  # Add color bar to show the values of z
plt.show()
```

```python
# Interactive Plot: *Use Matplotlib widgets (e.g., sliders or buttons) to update a graph dynamically.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Create initial data
x = np.linspace(0, 10, 1000)
amplitude_initial = 1
y = amplitude_initial * np.sin(x)

# Create the figure and axis
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.25)  # Adjust space for the slider

# Plot the initial data
line, = ax.plot(x, y, label="Sine Wave")

# Add a slider axis below the plot
axcolor = 'lightgoldenrodyellow'
ax_slider = plt.axes([0.25, 0.1, 0.65, 0.03], facecolor=axcolor)

# Create the slider to adjust the amplitude
slider = Slider(ax_slider, 'Amplitude', 0.1, 3.0, valinit=amplitude_initial)

# Define the update function for the slider
def update(val):
    amplitude = slider.val
    y = amplitude * np.sin(x)
    line.set_ydata(y)
    fig.canvas.draw_idle()  # Redraw the figure with the updated data

# Connect the slider to the update function
slider.on_changed(update)

# Display the plot
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Interactive Sine Wave Plot')
plt.show()
```
